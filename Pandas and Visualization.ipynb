```python
fl1 = fl.pivot('month', 'year', 'passengers')
fl1
```

```python
f, (axis1, axis2) = plt.subplots(2,1)
y_fl = fl1.sum()
years = pd.Series(y_fl.index.values)
years = pd.DataFrame(years)
flights = pd.DataFrame(pd.Series(y_fl.values))

years_dframe = pd.concat((years, flights), axis = 1)
years_dframe.columns = ['Years', 'Flights']

sns.barplot('Years', y= 'Flights', data = years_dframe, color = 'b', ax = axis1)
sns.heatmap(fl1, cmap = 'Blues', ax = axis2, cbar_kws = {'orientation': 'horizontal'})
```
